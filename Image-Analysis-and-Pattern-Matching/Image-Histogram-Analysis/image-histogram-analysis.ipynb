**فرمت فایل**

- JPEG (Joint Photographic Experts Group)
- پسوندهای رایج: `.jpg` یا `.jpeg`

---
**ساختار فایل JPEG**

JPEG از دو بخش کلی تشکیل شده است:

- **هدر (Header)**

  شامل متادیتا مانند رزولوشن تصویر، فضای رنگ ، جداول کوانتیزه سازی، الگوریتم فشرده سازی، اطلاعات (دوربین، تاریخ، GPS و ...)Exif

- **بیت‌ استریم (Bitstream)**

  داده‌های واقعی تصویر، فشرده‌سازی شده با الگوریتم
JPEG
---
**رزولوشن (Resolution)**

  با استفاده از کد پایتون در پایین نشان داده شده است.

---
**فشرده بودن یا نبودن**

JPEG ذاتاً فشرده‌سازی دارد و همانطور که در کتاب گفته شده از نوع lossy است یعنی در هنگام فشرده سازی بخشی از اطلاعات تصویر از بین می رود ولی این حذف به‌گونه‌ای انجام می‌شود که توسط چشم انسان کمتر قابل تشخیص است.

---
**الگوریتم فشرده سازی**

  مراحل فشرده سازی در این فرمت را بطور خلاصه می توان به شکل زیر در نظر گرفت:

- 1)تبدیل رنگ -> از RGB به YCbCr برای جدا کردن روشنایی از رنگ

- 2) تقسیم به بلاک های 8*8 -> تقسیم تصویر به بلوک‌های کوچک

- 3)تبدیل کسینوسی -> برای تبدیل داده‌ها به فضای فرکانس

- 4)کوانتیزه‌سازی -> کاهش دقت مقادیر (و حذف اطلاعات جزئی)

- 5)فشرده‌سازی نهایی -> معمولاً با Huffman coding
