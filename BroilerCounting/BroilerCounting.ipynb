```python
import zipfile
import os

zip_file_name = 'counting chickens.v4i.yolov8.zip'

with zipfile.ZipFile(zip_file_name, 'r') as zip_ref:
    zip_ref.extractall('yolo_dataset')

# Change the working directory to the dataset directory
os.chdir('yolo_dataset')
```

```python
from ultralytics import YOLO

# Load a pre-trained YOLOv8 model
# 'yolov8n.pt' is the nano version, which is fast and good for a start
# 'yolov8m.pt' is the medium version, more accurate but slower
model = YOLO('yolov8n.pt')

# Fine-tune the model on your custom dataset
results = model.train(data='data.yaml', epochs=50, imgsz=640)
```

```python
# Evaluate the model on the test dataset
metrics = model.val()
print(metrics.box.map50)  # Print mAP@0.5 metric
```

```python
# Run inference on a test image and show the results
# Replace 'test/images/your_test_image.jpg' with a path to one of your test images
test_image_path = 'test/images/your_test_image.jpg'
result_inference = model(test_image_path)
result_inference[0].show()
```

```python
# Import necessary libraries
from ultralytics import YOLO
from google.colab import files
from IPython.display import display, Image
from PIL import Image as PILImage
import io
import cv2
import numpy as np

# Load the trained model
model_path = '/content/yolo_dataset/runs/detect/train/weights/best.pt'
model = YOLO(model_path)

# Create file upload box
print("Please upload an image from your computer to test the model.")
uploaded_file = files.upload()

# Get the name of the uploaded file
if uploaded_file:
    uploaded_file_name = list(uploaded_file.keys())[0]

    # Read the image data and open it as a PIL Image
    image_data = uploaded_file[uploaded_file_name]
    image = PILImage.open(io.BytesIO(image_data))

    # Convert PIL Image to a format that OpenCV can use (numpy array)
    image_np = np.array(image)
    image_bgr = cv2.cvtColor(image_np, cv2.COLOR_RGB2BGR)

    # Run inference on the image with a confidence threshold
    results = model(image_bgr, conf=0.70)

    # Draw custom bounding boxes and labels
    for r in results:
        boxes = r.boxes
        for box in boxes:
            # Get bounding box coordinates
            x1, y1, x2, y2 = box.xyxy[0]
            x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)

            # Get confidence and class
            conf = float(box.conf[0])
            cls = int(box.cls[0])

            # Draw rectangle
            color = (200, 200, 200) # Green color
            thickness = 1
            cv2.rectangle(image_bgr, (x1, y1), (x2, y2), color, thickness)

            # Put label text
            label = f'{model.names[cls]}: {conf:.2f}'
            font = cv2.FONT_HERSHEY_SIMPLEX
            font_scale = 0.4  # Adjust font size here
            font_thickness = 2 # Adjust font thickness here
            text_size, _ = cv2.getTextSize(label, font, font_scale, font_thickness)
            cv2.rectangle(image_bgr, (x1, y1), (x1 + text_size[0], y1 - text_size[1] - 5), color, -1)
            cv2.putText(image_bgr, label, (x1, y1 - 5), font, font_scale, (0, 0, 0), font_thickness, cv2.LINE_AA)

    # Get the count of detected chickens
    chicken_count = len(results[0].boxes)

    # Convert the image back to RGB for display in Colab
    image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

    # Display the final image using IPython.display
    display(PILImage.fromarray(image_rgb))

    # Print the count----
    print(f"Number of chickens detected: {chicken_count}")
```
